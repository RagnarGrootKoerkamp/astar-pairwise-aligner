```python
import matplotlib.pyplot as plt
import numpy as np
import math

def ff(g): return g*g
def work_ng(s, g):
    sum = 0
    f = 1
    while f < s:
        sum += ff(f)
        f *= g
    sum += ff(f)
    return sum

L = 300
H = 3000

def test(g, f):
    return sum(f(s, g)/math.exp(0.01*s) for s in range(L, H))
    
g = np.arange(1.1, 3, 0.01)
b = [test(g, lambda s, g: ff(s)) for g in g]
t = [test(g, work_ng) for g in g]
plt.plot(g, t, 'b', g, [2*b for b in b], 'g', g, [4*b for b in b], 'r')
plt.ylim(ymin=0) 
plt.show()
```
